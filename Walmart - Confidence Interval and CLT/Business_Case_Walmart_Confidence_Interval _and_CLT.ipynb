### About Walmart 
Walmart is an American multinational retail corporation that operates a chain of supercenters, discount departmental stores, and grocery stores from the United States. Walmart has more than 100 million customers worldwide.

### Business Problem
The Management team at Walmart Inc. wants to analyze the customer purchase behavior (specifically, purchase amount) against the customerâ€™s gender and the various other factors to help the business make better decisions. They want to understand if the spending habits differ between male and female customers - <b>Do women spend more on Black Friday than men? (Assume 50 million customers are male and 50 million are female)</b>

```python
# Some utility functions

def draw_plot_numerical(column, bins = 15, color='c'):
    fig = plt.figure(figsize = (16, 6), dpi=100)

    fig1 = fig.add_subplot(2,1,1)
    sns.histplot(x=df[column], kde='reg', bins=bins, color=color)
    plt.axvline(df[column].mean(), color="g", label='Mean')
    plt.axvline(df[column].median(), color="black", label = 'Median')
    plt.axvline(df[column].mode()[0], color="r", label = 'Mode')
    plt.legend()
    plt.ylabel('')
    plt.xlabel('')

    fig.add_subplot(2,1,2, sharex=fig1)
    sns.boxplot(x=df[column], color=color)

    plt.suptitle(f'Analying {column} column', fontsize=20)
    plt.axvline(df[column].mean(), color="g", label='Mean')
    plt.axvline(df[column].median(), color="black", label = 'Median')
    plt.axvline(df[column].mode()[0], color="r", label = 'Mode')
    plt.xlabel('')
    plt.xticks(np.linspace(df[column].min(), df[column].max(), num=bins+1))
    plt.show()
    
def describe_and_find_outliers(column):
    print('-'*100)
    print(df[column].describe().to_string())
    
    Q1 = df[column].quantile(.25)
    Q3 = df[column].quantile(.75)
    IQR = Q3 - Q1

    right = df[df[column] > (Q3 + 1.5 * IQR)]
    left = df[df[column] < (Q1 - 1.5 * IQR)]
    
    print('-'*100)
    if(len(left) > 0):
        print('\nOutliers on left extreme:-\n')
        print(len(left))
    else:
        print('No outliers on left extreme')
        
    print('-'*100)
    if(len(right) > 0) :
        print('\nOutliers on right extreme:-\n')
        print(f'Total {len(right)} outliers which are greated than {Q3 + 1.5 * IQR}')
    else:
        print('No outliers on right extreme')
    
    print('-'*100)
    
    return left, right
```

# Confidence Interval and CLT
-> **Summed up customers behavior, insights and recommendations at the end of notebook**

```python
# Some utility functions

def bootstrap_data(data, sample, confidence_level, iteration=1000):
    
    bootstrap_means = np.empty(iteration)
    for i in range(iteration):
        bootstrap_sample = np.random.choice(data, size = sample, replace = True)
        bootstrap_means[i] = np.mean(bootstrap_sample)
        
    mean = np.mean(bootstrap_means)
    sigma = statistics.stdev(bootstrap_means)
    sem = stats.sem(bootstrap_means)
    
    lower_ci = np.percentile(bootstrap_means, (1 - confidence_level) / 2 * 100)
    upper_ci = np.percentile(bootstrap_means, (1 + confidence_level) / 2 * 100)
    
    return bootstrap_means, (mean, sigma, sem), (lower_ci, upper_ci)

def plotkde_ci(a, b, label_a, label_b, color_a, color_b, sample_size, confidence_levels, iteration=1000):
    
    for confidence_level in confidence_levels:
    
        a_data, a_info, a_ci = bootstrap_data(a, sample=sample_size, confidence_level=confidence_level)
        b_data, b_info, b_ci = bootstrap_data(b, sample=sample_size, confidence_level=confidence_level)

        plt.figure(figsize=(16, 8), dpi=100)

        sns.kdeplot(data = a_data, color = color_a, fill = True, linewidth=2)
        sns.kdeplot(data = b_data, color = color_b, fill = True, linewidth=2)

        plt.title(f"Sample Size: {sample_size}, Confidence level: {confidence_level}", fontsize=20)
        plt.xlabel('Purchase amount')

        plt.axvline(a_info[0], color = 'black', linestyle = 'dashdot', linewidth = 2,
                        label = (f"Î¼ ({label_a}):  {a_info[0]:.2f} \nSME ({label_a}): {a_info[2]:.2f}"))
        plt.axvline(a_ci[0], color = color_a, linestyle = 'solid', linewidth = 4,
                        label= (f"Lower Limit({label_a}): {a_ci[0]:.2f} \nUpper Limit({label_a}): {a_ci[1]:.2f}"))
        plt.axvline(a_ci[1], color = color_a, linestyle = 'solid', linewidth = 4)

        plt.axvline(b_info[0], color = 'black', linestyle = 'dashdot', linewidth = 2,
                       label = (f"Î¼ ({label_b}):  {b_info[0]:.2f} \nSME ({label_b}): {b_info[2]:.2f}"))
        plt.axvline(b_ci[0], color = color_b, linestyle = 'solid', linewidth = 4,
                        label=(f"Lower Limit({label_b}): {b_ci[0]:.2f} \nUpper Limit({label_b}): {b_ci[1]:.2f}"))
        plt.axvline(b_ci[1], color = color_b, linestyle = 'solid', linewidth = 4)

        plt.legend(bbox_to_anchor=(1.05, 1.05))
        plt.show()
```

### Do women spend more on Black Friday than men?

**Black Friday Sales - Insights:**

- Male customers spend significantly more than females during Black Friday sales.

**Black Friday Sales - Recommendations:**

- Tailor marketing and promotions to attract and retain male customers.
- Offer gender-specific deals and discounts to enhance engagement.
- Create a personalized shopping experience for both genders to maximize sales.

## Insights

1. **Customer Demographics:**
   - Around 80% of Walmart's customers fall within the age range of 18-50, with the majority (40%) aged between 26-35, indicating a significant market segment.
   - The customer base is predominantly male (75%) and largely single (60%).
   - City Category B houses the highest proportion of customers (42%), followed by Category C (31%), and Category A (27%).

2. **Purchasing Patterns:**
   - Product categories 1, 5, 8, and 11 experience the highest purchasing frequency, suggesting a focus on these categories in marketing strategies and product offerings.
   - The majority of purchases (86.31%) are made by customers between the ages of 26-50, highlighting the importance of targeting this age group in marketing and promotions.
   - Product categories 6, 7, and 15 are popular among customers, presenting opportunities to further promote and enhance offerings in these categories.
   - Customers residing in City Category C demonstrate higher spending per transaction, indicating potential for tailored marketing and promotional strategies for this city category.
   - Female customers have a preference for product category 5, suggesting targeted marketing or special promotions for this category to attract and retain female shoppers.

3. **Purchase Amount and Gender:**
   - Male customers tend to spend more than female customers, as indicated by higher average purchase amounts for males.
   - The confidence interval analysis consistently shows that male customers exhibit a wider spending range, further supporting the notion that males have a higher average transaction spending.

4. **Marital Status and Purchasing Behavior:**
   - The spending behavior of both married and unmarried customers is similar, as evidenced by overlapping confidence intervals, suggesting that marital status does not significantly affect purchase habits.
   - Approximately 59% of transactions are carried out by unmarried customers, highlighting the importance of targeting this demographic in marketing strategies.

5. **Occupation and City Category:**
   - Majority of customers fall under Occupation category 4, which presents an opportunity to tailor marketing strategies and product offerings to suit this occupation segment.
   - City Category B has the highest customer base, indicating the importance of focusing marketing efforts on this category, potentially through loyalty programs or special offers.

6. **Age Groups and Purchase Behavior:**
   - Different age groups exhibit overlapping purchase ranges, implying that age might not have a significant impact on purchase habits.
   - However, special attention could be given to the (0-17) and (51-55) age groups, as they exhibit slightly distinct purchasing patterns.

7. **Product Category Popularity:**
   - Product categories 5, 6, 7, and 15 are popular among customers, suggesting a need to ensure sufficient stock and targeted marketing for these categories to capitalize on their popularity.

## Recommendations

1. **Targeting Male Customers:**
   - Focus on retaining and attracting more male customers due to their higher spending compared to females.
   - Tailor marketing strategies and offers to cater specifically to male preferences and needs.

2. **Product Categories:**
   - Enhance sales by promoting products from categories 1, 5, 8, and 11, which exhibit high purchasing frequency.
   - Consider boosting products from less-purchased categories to diversify offerings and potentially increase sales.

3. **Acquisition of Unmarried Customers:**
   - Target unmarried customers who tend to spend more, devising marketing campaigns and offers to attract this segment.

4. **Targeting Age Group 18-45:**
   - Focus on acquisition strategies aimed at customers aged 18-45, who demonstrate higher spending compared to other age groups.

5. **City Category C Focus:**
   - Capitalize on the higher spending of male customers residing in City_Category C by promoting and selling more products in this category.

6. **Enhancing Female Shopping Experience:**
   - For Black Friday sales, offer a diverse range of products catering to female preferences to enhance their spending.
   - Implement loyalty programs and seek feedback to understand and address the specific needs of female customers.

7. **Improving Shopping Experience for Younger Generations:**
   - Introduce games and activities within the store to attract and engage the younger generation, enhancing the overall shopping experience.
   - Offer dedicated areas for children and teenagers, along with age-appropriate products and special discounts to boost sales in this demographic.

These recommendations focus on leveraging customer demographics and spending patterns to tailor strategies that drive growth, enhance customer engagement, and ultimately increase sales for Walmart.
