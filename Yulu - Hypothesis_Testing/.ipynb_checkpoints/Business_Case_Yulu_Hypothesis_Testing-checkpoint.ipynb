### About Yulu 
Yulu is Indiaâ€™s leading micro-mobility service provider, which offers unique vehicles for the daily commute. Starting off as a mission to eliminate traffic congestion in India, Yulu provides the safest commute solution through a user-friendly mobile app to enable shared, solo and sustainable commuting.

Yulu zones are located at all the appropriate locations (including metro stations, bus stands, office spaces, residential areas, corporate offices, etc) to make those first and last miles smooth, affordable, and convenient!

Yulu has recently suffered considerable dips in its revenues. They have contracted a consulting company to understand the factors on which the demand for these shared electric cycles depends. Specifically, they want to understand the factors affecting the demand for these shared electric cycles in the Indian market.

### Business Problem
The company wants to know:
- Which variables are significant in predicting the demand for shared electric cycles in the Indian market ?
- How well those variables describe the electric cycle demands ?

### Columns info:-
The company collected the transactional data of customers who purchased products from the Walmart Stores during Black Friday. The dataset has the following features:

- datetime: datetime
- season: season (1: spring, 2: summer, 3: fall, 4: winter)
- holiday: whether day is a holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
- workingday: if day is neither weekend nor holiday is 1, otherwise is 0.
- weather:
    - 1: Clear, Few clouds, partly cloudy, partly cloudy
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp: temperature in Celsius
- atemp: feeling temperature in Celsius
- humidity: humidity
- windspeed: wind speed
- casual: count of casual users
- registered: count of registered users
- count: count of total rental bikes including both casual and registered

```python
# Some utility functions

def draw_plot_numerical(column, bins = 15, color='c'):
    fig = plt.figure(figsize = (16, 6), dpi=100)

    fig1 = fig.add_subplot(2,1,1)
    sns.histplot(x=df[column], kde='reg', bins=bins, color=color)
    plt.axvline(df[column].mean(), color="g", label='Mean')
    plt.axvline(df[column].median(), color="black", label = 'Median')
    plt.axvline(df[column].mode()[0], color="r", label = 'Mode')
    plt.legend()
    plt.ylabel('')
    plt.xlabel('')

    fig.add_subplot(2,1,2, sharex=fig1)
    sns.boxplot(x=df[column], color=color)

    plt.suptitle(f'Analying {column} column', fontsize=20)
    plt.axvline(df[column].mean(), color="g", label='Mean')
    plt.axvline(df[column].median(), color="black", label = 'Median')
    plt.axvline(df[column].mode()[0], color="r", label = 'Mode')
    plt.xlabel('')
    plt.xticks(np.linspace(df[column].min(), df[column].max(), num=bins+1))
    plt.show()
    
def describe_and_find_outliers(column):
    print('-'*100)
    print(df[column].describe().to_string())
    
    Q1 = df[column].quantile(.25)
    Q3 = df[column].quantile(.75)
    IQR = Q3 - Q1

    right = df[df[column] > (Q3 + 1.5 * IQR)]
    left = df[df[column] < (Q1 - 1.5 * IQR)]
    
    print('-'*100)
    if(len(left) > 0):
        print('\nOutliers on left extreme:-\n')
        print(f'Total {len(left)} outliers which are lesser than {Q1 - 1.5 * IQR}')
    else:
        print('No outliers on left extreme')
        
    print('-'*100)
    if(len(right) > 0) :
        print('\nOutliers on right extreme:-\n')
        print(f'Total {len(right)} outliers which are greater than {Q3 + 1.5 * IQR}')
    else:
        print('No outliers on right extreme')
    
    print('-'*100)
    
    return left, right
```

# Hypothesis Testing

```python
# Setting significance value as 0.05 for all experiments
ALPHA = 0.05
```

```python
# Function for Normality tests - Histplot, QQplot and Shapiro-Wilk test

def test_normality(sample):
    
    # Visual Analysis
    sns.histplot(sample, kde = True)
    plt.title('Distribution of sample')
    
    # QQplot
    qqplot(sample, line = 's')
    plt.title('QQ plot of sample')
    plt.show()
    
    # Shapiro-Wilk test
    print("\nPerforming Shapiro-Wilk test for normality:-\n")
    print("Null Hypothesis: Given sample is normally distributed")
    print("Alternate Hypothesis: Given sample is not normally distributed")
    print(f'ALPHA (Significance value): {ALPHA}\n')
    
    statistic, p_value = shapiro(sample) 
    
    print(f'Statistic: {statistic}, p-value: {p_value}')
    if p_value < ALPHA:
        print("Reject null hypothesis, given sample is not normal distributed.")
    else:
        print("Fail to reject null hypothesis, given sample is normal distributed.")
        
        
        
# Function for Levene test for equal variances
def levene_test(labels, *samples):
    
    # Visual Analysis
    fig = plt.subplots(nrows = 1,ncols = 2, figsize = (15, 5))
    plt.suptitle(f"Comparing variance of samples", fontsize=20)
    
    plt.subplot(1,2,1)
    for sample in samples:
        sns.kdeplot(data = sample, fill = True, linewidth=2)
    
    plt.subplot(1,2,2)
    sns.boxplot(pd.DataFrame({labels[idx]: sample.values for idx, sample in enumerate(samples)}))
    plt.show()
    
    print("\nPerforming Levene test for equal variances:-\n")
    print("Null Hypothesis: All samples have equal variances")
    print("Alternate Hypothesis: Atleast one sample has different variance from rest")
    print(f'ALPHA (Significance value): {ALPHA}\n')

    result = levene(*samples)
    statistic, p_value = result.statistic, result.pvalue
    
    print(f'Statistic: {statistic}, p-value: {p_value}')
    if p_value < ALPHA:
        print("Reject null hypothesis, atleast one sample has different variance from rest.")
    else:
        print("Fail to reject null hypothesis, all samples have equal variance.")
```

## Two-sample T-Test

The two-sample t-test (also known as the independent samples t-test) is a method used to test whether the unknown population means of two groups are equal or not.

    T-Statistic = (1_mean - s2_mean) / np.sqrt((s1_std**2 / s1_sample_size) + (s2_std**2 / s2_sample_size))
    
    ttest_ind(samples_1, samples_2, alternative={'two-sided', 'less', 'greater'})
    
    ttest_ind_from_stats(mean1, std1,  nobs1, mean2, std2, nobs2, alternative={'two-sided', 'less', 'greater'})

<b>Assumptions:-</b>
- Data values must be independent. Measurements for one observation do not affect measurements for any other observation.
- Data in each group must be obtained via a random sample from the population.
- Data in each group are normally distributed.
- Data values are continuous.
- The variances for the two independent groups are equal.

```python
# Function for T-test for the means of *two independent* samples of scores

def t_test_independent(labels, sample_a, sample_b):
    
    # Visual Analysis
    plt.figure(figsize=(8, 6))
    plt.title(f"Comparing variance of samples", fontsize=20)
    for idx, sample in enumerate([sample_a, sample_b]):
        ax = sns.kdeplot(data = sample, fill = True, linewidth=2)
        plt.axvline(np.mean(sample), linestyle = 'dashdot', linewidth = 2,
                    label = f"Mean ({labels[idx]}): {np.mean(sample):.2f}",
                    color = next(ax._get_lines.prop_cycler)['color'])
    plt.legend(bbox_to_anchor=(1.7, 1.05))
    plt.show()
    
    print("\nPerforming T-test for the means of two independent samples:-\n")
    print("Null Hypothesis: Given samples have same means")
    print("Alternate Hypothesis: Given samples have different means")
    print(f'ALPHA (Significance value): {ALPHA}\n')
    
    result = ttest_ind(sample_a, sample_b)
    statistic, p_value, dof = result.statistic, result.pvalue, result.df
    
    print(f'Statistic: {statistic}, p-value: {p_value}')
    if p_value < ALPHA:
        print("Reject null hypothesis, given samples have different means.")
    else:
        print("Fail to reject null hypothesis, given samples have same means.")
```

## ANOVA (One-Way) (Analysis of Variance)

The one-way ANOVA tests the null hypothesis that two or more groups have the same population mean.  The test is applied to samples from two or more groups, possibly with differing sizes.

    F-statistic = Variance between the groups / Variance within each group
    
    f_oneway(sample_1, sample_2, sample_3.....sample_n)

- How many T-Tests for n different samples - nC2. To speed up, we use one-way anova for more than 2 groups.
- Null hypothesis -> There is no difference in mean of samples
- Alternate hypothesis -> Atleast one sample mean is different from rest

<b>Assumptions:-</b>
- Data should be normally distributed
- Data should be independent accross each record
- Variance of different groups should be same

<b>If assumptions are not met, we can use Kruskal Wallis Test</b>.

```python
# Function for One-way ANOVA test

def anova_test(labels, *samples):
    
    # Visual Analysis
    plt.figure(figsize=(8, 6))
    plt.title(f"Comparing variance of samples", fontsize=20)
    for idx, sample in enumerate(samples):
        ax = sns.kdeplot(data = sample, fill = True, linewidth=2)
        plt.axvline(np.mean(sample), linestyle = 'dashdot', linewidth = 2,
                    label = f"Mean ({labels[idx]}): {np.mean(sample):.2f}",
                    color = next(ax._get_lines.prop_cycler)['color'])
    plt.legend(bbox_to_anchor=(1.7, 1.05))
    plt.show()
    

    print("\nPerforming ANOVA test for the means of independent samples:-\n")
    print("Null Hypothesis: Given samples has same means")
    print("Alternate Hypothesis: Atleast one sample has different mean from rest")
    print(f'ALPHA (Significance value): {ALPHA}\n')
    
    result = f_oneway(*samples)
    statistic, p_value = result.statistic, result.pvalue
    
    print(f'Statistic: {statistic}, p-value: {p_value}')
    if p_value < ALPHA:
        print("Reject null hypothesis, atleast one sample has different mean from rest.")
    else:
        print("Fail to reject null hypothesis, given samples have same means.")
```

## Chi-Squared Test

<b>Chi-Squared Independence of Features test</b>

A hypothesis test designed to test for a statistically significant relationship between nominal and ordinal variables organized in a bivariate table. In other words, it tells us whether two variables are independent of one another.

    X2 (Chi-Statistic) = Summation over -> (Observed - Expected)**2 / Expected
    
    chi2_contingency(observed_contingency_table)
    
<b>Assumptions of Chi2 test</b>
- Variables are categorical
- Observations are independent
- Each cell is mutually exclusive
- Expected value in each cell is greater than 5 (at least in 80% of cells)

```python
# Function for Chi-square test of independence of variables in a contingency table

def chi_square_test(contingency_table):
    
    # Visual Analysis
    plt.figure(figsize=(20, 5), dpi=100)
    sns.heatmap(data=contingency_table, annot=True, linewidths=1, cmap='Greens',
                annot_kws={"fontsize":18}, fmt='.0f')
    plt.show()

    print("\nPerforming Chi-square test of independence of variables in a contingency table:-\n")
    print("Null Hypothesis: Observed values in the contingency table are independent")
    print("Alternate Hypothesis: Observed values in the contingency table are dependent")
    print(f'ALPHA (Significance value): {ALPHA}\n')

    result = chi2_contingency(contingency_table)
    statistic, p_value, dof, expected = result.statistic, result.pvalue, result.dof, result.expected_freq

    print(f'Statistic: {statistic}, p-value: {p_value}')
    if p_value < ALPHA:
        print("Reject null hypothesis, observed values in the contingency table are dependent.")
    else:
        print("Fail to reject null hypothesis, observed values in the contingency table are independent.")
```

```python
# Utility functions

def get_all_category_samples(column, sample_size, apply_log_transformation = False):
    if sample_size > df[column].value_counts().min():
        raise Exception(f"Sample size cannot be more than {df[column].value_counts().min()}")

    output = {}
    for category in df[column].unique():
        sample = df[df[column] == category]['count'].sample(sample_size)
        if apply_log_transformation:
            sample = np.log(sample)
        output[category] = sample
        
    return output


def test_normality_categories(category_samples):
    for category, sample in category_samples.items():
        print('='*100)
        print(f"\nTest for normality of sample of category - {category}")
        test_normality(sample)
```

   - Fall season records the highest rental rate at approximately 30.7%, followed by summer at 28.2%.
   - Winter season has a rental rate of about 26.1%, while spring records the lowest at 15%.

2. **Weather Impact:**
   - Weather conditions significantly affect bike rentals.
   - Weather condition 1 (Clear) experiences the highest rental rate, while condition 4 (e.g., rain, snow) has an exceptionally low rental rate.
   - Humidity below 20 and temperature below 10 or above 35 correlate with lower bike rentals.

3. **Day and Time Patterns:**
   - On working days, 68.6% of cycles are rented, with no statistically significant difference compared to non-working days.
   - Peak rental hours are around 5 PM, 6 PM, and 8 AM, indicating commuting patterns.
   - Increasing trend in rentals from 5 AM to 8 AM and decreasing trend from 5 PM to 11 PM.

4. **Monthly Trends:**
   - Highest average hourly bike rentals in June, July, and August (summer months).
   - Lowest in January, February, and March (winter months).
   - Increasing trend from February to June and decreasing trend from October to December.

5. **Correlations:**
   - Positive correlation between temperature and bike rentals (0.39), indicating increased rentals with higher temperatures.
   - Negative correlation between humidity and bike rentals, suggesting lower rentals during high humidity conditions.
   - Weak positive correlation between windspeed and bike rentals, indicating a niche preference for cycling in windy conditions.

6. **Booking Preferences:**
   - Clear weather attracts about 2/3 of the rides, showing a preference for booking Yulu in clear conditions.
